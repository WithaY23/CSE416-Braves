```python
import numpy as np
import pandas as pd
import json
import csv
```

# GUI-17 Population Percentages Box & Whisker

Calculate current Oregon district plan's populations

```python
base_path = "/content/drive/MyDrive/Preprocessing/Preprocessing11"
```

```python
or_path = f"{base_path}/or_combined.csv"
or_current_plan_df = pd.read_csv(or_path)
or_current_plan_df

or_data = [
    ["district", "total", "white", "black", "hispanic"],
]

districts = [[1, 0, 0, 0, 0], [2, 0, 0, 0, 0], [3, 0, 0, 0, 0], [4, 0, 0, 0, 0], [5, 0, 0, 0, 0], [6, 0, 0, 0, 0]]
for _, row in or_current_plan_df.iterrows():
  district = -1
  match row["district"]:
    case 1:
      district = districts[0]
    case 2:
      district = districts[1]
    case 3:
      district = districts[2]
    case 4:
      district = districts[3]
    case 5:
      district = districts[4]
    case 6:
      district = districts[5]
    case _:
      raise Exception("Error: Oregon should have 6 districts, or district number data column should be labeled \"district\" ")

  district[1] += row["total"]
  district[2] += row["white"]
  district[3] += row["black"]
  district[4] += row["hispanic"]

for district_data in districts:
  or_data.append(district_data)

with open(f"{base_path}/or_current_districts_pops.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(or_data)
```

Total Oregon state populations

```python
or_path = f"{base_path}/or_current_districts_pops.csv"
or_current_pops_df = pd.read_csv(or_path)

or_totals = or_current_pops_df.iloc[:, 1:].sum().to_dict()

or_totals
```

Rank districts based on Hispanic population and calculate percentages over total

```python
or_current_pops_df = or_current_pops_df.sort_values(by=["hispanic"])
or_enacted_vals = []

for _, row in or_current_pops_df.iterrows():
  or_enacted_vals.append((row["hispanic"] / row["total"]).item())

print(or_enacted_vals)
```

Box and Whiskers

```python
df_path = f"{base_path}/regular_bw.json"
df = pd.read_json(df_path)

print(f"Number of Oregon raceblind ensembles: {len(df["oregon"]["raceblind"])}")
for i in range(len(df["oregon"]["raceblind"])):
  print(f"Number of Oregon raceblind plans in index {i}: {len(df["oregon"]["raceblind"][i])}")

print(f"Number of Oregon vra ensembles: {len(df["oregon"]["vra"])}")
for i in range(len(df["oregon"]["vra"])):
  print(f"Number of Oregon raceblind plans in index {i}: {len(df["oregon"]["vra"][i])}")

print(f"Number of South Carolina raceblind ensembles: {len(df["south_carolina"]["raceblind"])}")
for i in range(len(df["south_carolina"]["raceblind"])):
  print(f"Number of South Carolina raceblind plans in index {i}: {len(df["south_carolina"]["raceblind"][i])}")

print(f"Number of South Carolina vra ensembles: {len(df["south_carolina"]["vra"])}")
for i in range(len(df["south_carolina"]["vra"])):
  print(f"Number of South Carolina vra plans in index {i}: {len(df["south_carolina"]["vra"][i])}")
```

Calculate Percentages for Race-blind Ensembles

```python
states = ["oregon", "south_carolina"]

for state in states:
  if state == "oregon":
    enacted_vals = or_enacted_vals
    total = or_totals["hispanic"]
  else:
    enacted_vals = sc_enacted_vals
    total = sc_totals["black"]

  for ensemble_num in range(0, len(df[state]["raceblind"])):
    rank_summaries = []
    num_plans = len(df[state]["raceblind"][ensemble_num])
    num_districts = len(df[state]["raceblind"][0][0])

    for dist in range(num_districts):
        pcts_in_rank = []
        for n in range(num_plans):
          pcts_in_rank.append(df[state]["raceblind"][ensemble_num][n][dist] / total)

        population_percent_data = {
          "districtRank": dist + 1,
          "min": round(np.min(pcts_in_rank), 3),
          "q1": round(np.quantile(pcts_in_rank, 0.25), 3),
          "median": round(np.median(pcts_in_rank), 3),
          "q3": round(np.quantile(pcts_in_rank, 0.75), 3),
          "max": round(np.max(pcts_in_rank), 3),
          "enactedValue": round(enacted_vals[dist], 3)
        }

        rank_summaries.append(population_percent_data)

    with open(f"{base_path}/output/GUI17/GUI17_{toStateCode(state)}_rb{ensemble_num}_{num_plans}plans.json", "w") as f:
      json.dump(rank_summaries, f, indent=2, default=float)
```

# GUI-21 Minority Effectiveness Box & Whisker

```python
effectiveness_path = f"{base_path}/min_eff_bw.json"
effectiveness_df = pd.read_json(effectiveness_path)

print(effectiveness_df)
print(f"Number of Oregon race blind ensembles: {len(effectiveness_df["oregon"]["raceblind"])}")
print(f"Oregon raceblind ensembles: {effectiveness_df["oregon"]["raceblind"]}")
print(f"Oregon Hispanic rb ensemble 2: {(effectiveness_df["oregon"]["raceblind"][1]["hispanic"])}")
print(f"Oregon White rb ensemble 2: {(effectiveness_df["oregon"]["raceblind"][1]["white"])}")
print(f"Number of Oregon vra ensembles: {len(effectiveness_df["oregon"]["vra"])}")
print(f"Oregon vra ensembles: {effectiveness_df["oregon"]["vra"]}")
print(f"Oregon Hispanic vra ensemble 1: {(effectiveness_df["oregon"]["vra"][0]["hispanic"])}")

print(f"Number of South Carolina race blind ensembles: {len(effectiveness_df["south_carolina"]["raceblind"])}")
print(f"South Carolina raceblind ensembles: {effectiveness_df["south_carolina"]["raceblind"]}")
print(f"South Carolina Black rb ensemble 2: {(effectiveness_df["south_carolina"]["raceblind"][1]["black"])}")
print(f"South Carolina White rb ensemble 2: {(effectiveness_df["south_carolina"]["raceblind"][1]["white"])}")
print(f"Number of South Carolina vra ensembles: {len(effectiveness_df["south_carolina"]["vra"])}")
print(f"South Carolina vra ensembles: {effectiveness_df["south_carolina"]["vra"]}")
print(f"South Carolina Black vra ensemble 1: {(effectiveness_df["south_carolina"]["vra"][0]["black"])}")
```

Calculate Minority Effectiveness Box Whisker Statistics for Race Blind Ensembles

```python
for state in states:
  if state == "oregon":
    enacted_minority_effective = 0
    groups = ["hispanic", "white"]
  else:
    enacted_minority_effective = 1
    groups = ["black", "white"]
  group_summaries = []
  num_ensembles = len(effectiveness_df[state]["raceblind"])
  num_potential_dists = len(effectiveness_df[state]["raceblind"][0]["white"])

  for group in groups:
    for ensemble_num in range(num_ensembles):

      num_plans = sum(effectiveness_df[state]["raceblind"][ensemble_num][group].values())
      effective_district_counts = []

      for i in range(num_potential_dists):
        effective_district_counts.extend([i] * effectiveness_df[state]["raceblind"][ensemble_num][group].get(str(i)))

      data = {
        "key": group if group != "hispanic" else "latino",
        "label": group[0].upper() + group[1:] if group != "hispanic" else "Latino",
        "enactedValue": enacted_minority_effective
      }

      race_blind_summary = {
        "min": np.min(effective_district_counts),
        "q1": np.quantile(effective_district_counts, 0.25),
        "median": np.median(effective_district_counts),
        "q3": np.quantile(effective_district_counts, 0.75),
        "max": np.max(effective_district_counts)
      }

      data["raceBlindSummary"] = race_blind_summary

      with open(f"{base_path}/output/GUI21/GUI21_{toStateCode(state)}_rb{ensemble_num}_{group}_{num_plans}plans.json", "w") as f:
        json.dump(data, f, indent=2, default=float)
```
